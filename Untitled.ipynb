```python
def dataCollector():
    equityListPd = pd.DataFrame(equityList.find({'Security Name':'Active','State':"A"}))
    collectionTime=2000
    basicSample=30
    cnt=0
    my_time = datetime.now()

    for rows in equityListPd['Issuer Name'].iteritems():
        print(cnt,rows[1],end="-")
        cnt= cnt + 1
        if stockWarehouse.find({'Symbol':rows[1]}).count() > 1 :
            for j in stockWarehouse.find({'Symbol':rows[1]},{'_id':0,'Date':1}).sort('Date',-1).limit(1) :
                collectionTime=((my_time - j['Date']).days - 1)
                fromDate=j['Date']
        iters=int(collectionTime/basicSample)
        if (collectionTime != 0):
            if collectionTime < 2000:
                basicSample=collectionTime
                iters=1
            dt=datetime.now()
            print("Iterations:",iters,end="|")
            for i in range(0,iters):

                startDate=dt - timedelta(days=basicSample)
                if( startDate <= fromDate):
                    startDate = (fromDate + timedelta(days=1))
                endDate=dt 
                print(i,"Collecting into between",startDate.date(),"-", endDate.date(),"@",datetime.now(),end="|")
                df=pd.DataFrame(get_history(symbol=rows[1] , start= startDate,end=endDate))
                df.reset_index(inplace=True)
                df['Date']=pd.to_datetime(df['Date'])
                dfLength=(len(df.index))
                if  dfLength > 0 :
                    data_dict=df.to_dict("records")
                    stockWarehouse.insert_many(data_dict)
                    tmpdf=pd.DataFrame(stockWarehouse.find({"Symbol":rows[i],"Date":{"$lte":endDate,"$gte":startDate}}))
                    print(df.index.size,tmpdf.index.size,end="|")

                else :
                    print("Marking as Inactive")
                    equityList.update_many({'Issuer Name':rows[1]},{'$set': {'Security Name':'Inactive'}})
                    break;
                dt=(startDate - timedelta(days=1))
```

```python
def markInactiveStocks():
    reference=int(stockWarehouse.find({'Symbol':'SBIN'}).count()) - 100
    print(reference)
    tmp_count=0
    for i in equityList.find({},{'Issuer Name':1}):
        #print(i['Issuer Name'])
        tmp_count=(int(stockWarehouse.find({'Symbol':i['Issuer Name']}).count()))
        if tmp_count >= reference:
            print(i['Issuer Name'],end='|')
            equityList.update_many({'Issuer Name':i['Issuer Name']},{'$set': {'State':"A"}})
```

```python
#result=pd.DataFrame(columns=['Symbol','FHP','FLP','exit1','exit2','exit3','exit4','exit5','exit6'])
result=[]
equityList=db["stocklist"]
stockWarehouse=db["stockWarehouse"]

#########################################
# Get Stock List 
#########################################
equityListdf=pd.DataFrame(equityList.find({'Security Name':'Active','State':"A"}))

#########################################
# Get Reference Date 
#########################################
for dt in stockWarehouse.find({'Symbol':'SBIN'},{'_id':0,'Date':1}).sort('Date',-1).limit(1):
    refdate=dt # Set Reference Data as last date when SBI Stock Data was update
print(refdate)

#########################################
# Get Suggestions 
#########################################

for symbol in equityListdf['Issuer Name']    :
    for maxdate in stockWarehouse.find({'Symbol':symbol},{'_id':0,'Date':1}).sort('Date',-1).limit(1):
        if maxdate == refdate:  # Check if database has latest sttock info by comparing aginst SBI
            df=pd.DataFrame(stockWarehouse.find({'Symbol':symbol}).sort('Date',-1).limit(300))# Get top 300 row of stock in iteration
            df.set_index('Date',inplace=True)
            df.sort_index(inplace=True)
            df['EWM25']=df['Close'].ewm(span=25,adjust=False).mean()
            df['EWM50']=df['Close'].ewm(span=50,adjust=False).mean()
            df['EWM100']=df['Close'].ewm(span=100,adjust=False).mean()
            df['EWMDIFF']=df['EWM25']-df['EWM50']
            df['EWMDIFFpct']=df['EWMDIFF'].pct_change(periods=1)
            df['EWMDIFFpct'].fillna(0)
            trendCompare=df.iloc[-1]
            lastClose=df['Close'].iloc[-1]
            lastHigh=df['High'].iloc[-1]
            lastLow=df['Low'].iloc[-1]
            pivot=(lastClose+lastHigh+lastLow)/3
            resistence1=(2*pivot) - lastLow
            support1=(2*pivot) - lastHigh
            resistence2=pivot + resistence1-support1
            support2=pivot - resistence1-support1
            if trendCompare['EWM100'] < trendCompare['EWM50'] and trendCompare['EWMDIFFpct'] > .1 and trendCompare['Close'] < 1500 and trendCompare['Close'] > 50:   # Slow EWMA is less Than Fast EWMA indicating uptrend start
                crossOverCount=0
                i=df.shape[0]-1
                FLP=0
                FHP=0
                diff=0
                while i > 2:
                   # print(i,end="-")
                    difftoday=df['EWM25'].iloc[i]-df['EWM50'].iloc[i]
                    diffyesterday=df['EWM50'].iloc[i-1]-df['EWM25'].iloc[i-1]
                    if difftoday >= 0 and  diffyesterday > 0 and crossOverCount==0:  # find cutover for last lower peak
                        FLP=round(df['EWM25'].iloc[i])
                        #FLP=df['Low'].min()
                        Daysincetrendchange=300 - i
                        crossOverCount=1
                    if difftoday <= 0 and  diffyesterday < 0 and crossOverCount==1:  # find cutover for last Higher peak
                        FHP=round(df['EWM25'].iloc[i])
                        #FHP=df['High'].max()
                        crossOverCount=2
                        break
                    i=i-1
                if FLP != 0 and FHP !=0:
                    diff=FHP-FLP
                    #if (lastclose )
                    exit1=round(FLP + (diff * .25))
                    exit2=round(FLP + (diff * .38))
                    exit3=round(FLP + (diff * .5))
                    exit4=round(FLP + (diff * .61))
                    exit5=round(FLP + (diff * 1))
                    exit6=round(FLP + (diff * 1.68))
                #print(symbol,FHP,FLP,exit1,exit2,exit3,exit4,exit5,exit6)
                result.append([symbol,FHP,FLP,lastClose,lastHigh,lastLow,pivot,resistence1,support1,resistence2,support2,Daysincetrendchange,exit1,exit2,exit3,exit4,exit5,exit6])
                #print(result)
                
resultdf=pd.DataFrame(result,columns=['Symbol','FHP','FLP','lastclose','lastHigh','lastLow','pivot','resistence1','support1','resistence2','support2','Daysincetrendchange','exit1','exit2','exit3','exit4','exit5','exit6'])
#resultdf['exit1']
print(resultdf)
resultdf.to_csv("C:\\Users\\khanna's\\Documents\\Stock Trend Analysis\\result.csv")

                
                
                
                
```
